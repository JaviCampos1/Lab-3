```python
locations = {
    "admu":{
        "name":"Ateneo de Manila University",
        "latitude":14.6394557,
        "longitude":121.0758975        
    },
    "upd":{
        "name":"University of the Philippines-Diliman Campus",
        "latitude":14.6538519,
        "longitude":121.0663348
    },
    "dlsu":{
        "name":"De La Salle University-Taft",
        "latitude":14.5647346,
        "longitude":120.9909876

    }, 
}

legs = {
     ("upd","admu"):{
         "travel_time_mins":10
     },
     ("admu","dlsu"):{
         "travel_time_mins":35
     },
     ("dlsu","upd"):{
         "travel_time_mins":55
     }
}

def get_immediate_destination(source):
    routes = legs.keys()
    return [chosen_route[1] for chosen_route in routes if chosen_route[0] == source][0]

def eta(source, destination):
    immediate_destination = get_immediate_destination(source)
    if immediate_destination == destination:
        return legs[(source, destination)]["travel_time_mins"]
    else:
        immediate_destination2 = get_immediate_destination(immediate_destination)
        return legs[(source, immediate_destination)]["travel_time_mins"] + legs[(immediate_destination, immediate_destination2)]["travel_time_mins"]

if __name__ == "__main__":
    source = input("Enter source ")
    destination = input("Enter destination ")


    print (eta(source, destination))
    
```
